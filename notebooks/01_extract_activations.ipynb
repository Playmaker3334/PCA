```python
from src.utils import ACTIVATIONS, load_npz
for f in sorted(ACTIVATIONS.iterdir()):
    d = load_npz(f)
    keys = [k for k in d.files if k.startswith('layer_')]
    n = len(d['prompts']) if 'prompts' in d.files else 0
    shapes = {k: d[k].shape for k in keys}
    print(f'{f.name}: n={n} shapes={shapes}')
```
